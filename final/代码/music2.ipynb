## 一、探究性别

## 五、探究相关性

### 1.所有数据相关性

```python
import seaborn as sns
df
```

```python
#年龄为0不具有参考价值，将年龄不为0的用户提取出来，将性别不确定的用户提取出来
fl_df= df[(df['user_age'] != 0) & ((df['user_gender'] == 1) | (df['user_gender'] == 2))]
fl_df
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties

data = pd.DataFrame()
data['gender'] = fl_df['user_gender'].apply(lambda x: 1 if x==2 else 0)

# 标准化处理
data['age'] = (fl_df['user_age'] - np.mean(fl_df['user_age'])) / np.std(fl_df['user_age'])
data['praise'] = (fl_df['praise'] - np.mean(fl_df['praise'])) / np.std(fl_df['praise'])
data['day'] = (fl_df['day'] - np.mean(fl_df['day'])) / np.std(fl_df['day'])

# 计算相关系数
corr = data.corr()
corr = corr[['age']]
corr = corr.drop('age')

# 取绝对值排序
corr = corr.apply(lambda x: abs(x))
corr = corr.sort_values('age', ascending=False)

# 绘制特征关系图
sns.heatmap(data.corr(), cmap='coolwarm', annot=True, fmt='.2f', xticklabels=1, yticklabels=1, linewidths=0.5)
plt.xticks(rotation=90, fontsize=12)
plt.yticks(rotation=0, fontsize=12)
plt.title('Feature Correlation Heatmap')
plt.tight_layout()
plt.show()
```

探究年龄与获赞数量关系

```python
sns.scatterplot(x='user_age', y='praise', data=fl_df)

plt.title('relationship of age and praise')
plt.xlabel('user_age')
plt.ylabel('counts')

# 显示图形
plt.show()
```

可以发现评论区用户年龄在20岁左右评论获赞数较多

```python
sns.pairplot(data=fl_df, hue='comment_id', vars=['user_age', 'user_gender', 'praise', 'month'])
plt.title('Pairplot')
plt.show()
```

探究月份对应评论数统计

```python
count_per_month = fl_df['month'].value_counts().sort_index()
plt.plot(count_per_month.index.astype(str), count_per_month.values, marker='o')
plt.xlabel('Month')
plt.ylabel('Count')
plt.title('Count per Month')
plt.xticks(rotation=45)
plt.show()
```

发现在12月份及8月份评论数量较多

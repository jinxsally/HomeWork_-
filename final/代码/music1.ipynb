```python
import pandas as pd

raw_data = pd.read_csv("D:/songs/total.csv")
raw_data
```

## 1.缺失值处理

```python

raw_data.drop_duplicates(subset=['comment_id'],inplace=True)
data=raw_data
data
```

```python
#查看各列缺失值
data.isna().sum()
```

```python
#删除comment_id为0的列
data.dropna(subset=['comment_id'],inplace=True)
data
```

```python
#再次查看缺失值
data.isna().sum()
```

```python
#将本文数据用空字符串填充空值
data.loc[:,'user_introduce'] = data['user_introduce'].fillna('').astype(str)
data
```

```python
data.isna().sum()
```

## 2.数据变换

```python
data.loc[:,'praise'] = data['praise'].astype(int)
data
```

```python
#转换为时间戳
data.loc[:, 'date'] = pd.to_datetime(data['date'])
data
```

```python
data.to_csv('D:/total_已处理.csv', index=False)
```
